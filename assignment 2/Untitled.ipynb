```python
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
```

```python
def train_test_split(x,y,test_size):
    rand = np.random.rand(x.shape[0])
    split = rand < np.percentile(rand,100 - (100*test_size))
    x_train = x[split]
    x_test = x[~split]
    y_train = y[split]
    y_test = y[~split]
    return x_train, x_test, y_train, y_test
```

```python
def sign(w,x):
    sum =0
    for i in range(len(x)):
        sum += w[i]+x[i]
    if sum == 0:
        return 0
    elif sum<0:
        return -1
    else :
        return 1
```

```python
def test(w,x):
    sum =0
    for i in range(len(x)):
        sum += w[i]+x[i]
    return sum
    
```

```python
def transform(x):
    return np.append(x,[1])
```

```python
def select(y,n):
    arr = np.array([])
    for i in y:
        if i == n:
            arr = np.append(arr,[1])
        else:
             arr = np.append(arr,[-1])
    return(arr)      
    
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.2)
```

```python
w=[0] *(4+1)
y = select(y_train,0)
print(y)
while True:
    m = 0
    for i in range(len(x_train)):
        x = transform(x_train[i])
        a = int(y[i]) ** (sign(w,x))
        print(i,"---",x,"sum: ",test(w,x),"sign:",sign(w,x),"y:",int(y[i]),"a:",a)
        if int(y[i]) ** (sign(w,x)) <= 0:
            print("       w:",w,"+",int(y[i])*x)
            w = w+(int(y[i])*x)
            print("       w:",w)
            m = m+1
    if m == 0:
        break
```
